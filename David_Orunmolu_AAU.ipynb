```python
#Join train and test set
newtrain = train_dataset.shape[0]
newtest = test_dataset.shape[0]

combined_dataset = pd.concat([train_dataset, test_dataset],ignore_index=True, sort=True)
```

```python
combined_dataset.shape
```

```python
cols_2_label_enc = ['State_Of_Origin', 'Division', 'No_of_previous_employers']
```

```python
from sklearn.preprocessing import LabelEncoder

#create an instance
lb = LabelEncoder() #create an object before using

for cat in cols_2_label_enc:
    lb.fit(combined_dataset[cat]) #Fit the column by learning what values to use for encoding
    combined_dataset[cat] = lb.transform(combined_dataset[cat]) #transform the column with the learnt value
```

```python
combined_dataset.head()=
```

ONE HOT ENCODING THE REST COLUMNS IN THE DATASET

```python
combined_dataset = pd.get_dummies(combined_dataset)
```

```python
combined_dataset.head()
```

```python
#get back train and test sets
train_dataset = combined_dataset[0:newtrain]
test_dataset = combined_dataset[newtrain:]
```

```python
# Getting our target
target = train_dataset['Promoted_or_Not']

train_dataset.drop('Promoted_or_Not', axis=1, inplace=True)
test_dataset.drop('Promoted_or_Not', axis=1, inplace=True)
```

```python
#Split dataset into the training set and test test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_dataset, target, test_size=0.25)
```
